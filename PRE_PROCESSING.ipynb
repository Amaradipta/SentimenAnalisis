```python
import re
import string
import nltk

def remove_URL(tweet):
  url = re.compile(r'https?://\s+/www\.\s+')
  return url.sub(r'', tweet)

def remove_html(tweet):
  html = re.compile(r'<.*?>')
  return html.sub(r'', tweet)

def remove_emoji(tweet):
  emoji_patern = re.compile("["
     u"\U0001F600-\U0001F64F"
     u"\U0001F300-\U0001F5FF"
     u"\U0001F680-\U0001F6FF"
     u"\U0001F1E0-\U0001F1FF"
                      "]+", flags=re.UNICODE)
  return emoji_patern.sub(r'', tweet)

def remove_numbers(tweet):
    tweet = re.sub(r'\d+', '', tweet)
    return tweet

def remove_symbols(tweet):
  tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)
  return tweet

df['cleansing'] = df['review'].apply(lambda x: remove_URL(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_html(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_emoji(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_symbols(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_numbers(x))

df.head(1000)
```

```python
df.to_csv('cleansing_imdb.csv',encoding='utf8', index=False)
```

```python
def case_folding(text):
  if isinstance(text, str):
    lowercase_text = text.lower()
    return lowercase_text
  else:
    return text

df['case_folding'] = df['cleansing'].apply(case_folding)

df.head(1000)
```

```python
df.to_csv('casefolding_imdb.csv',encoding='utf8', index=False)
```

```python
def tokenize(text):
  tokens = text.split()
  return tokens

df['tokenize'] = df['case_folding'].apply(tokenize)

df.head(1000)
```

```python
df.to_csv('tokenize_imdb.csv',encoding='utf8', index=False)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(text):
  return [word for word in text if word not in stop_words]

df['Filtering/stopword removal'] = df['tokenize'].apply(lambda x: remove_stopwords(x))

df.head(1000)
```

```python
df.to_csv('stopword_imdb.csv',encoding='utf8', index=False)
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='english')

def stem_text(text):
  return [stemmer.stem(word) for word in text]

df['stemming_data'] = df['Filtering/stopword removal'].apply(lambda x: ' '.join(stem_text(x)))

df['comments'] = df['Filtering/stopword removal'].apply(lambda x: ' '.join(stem_text(x)))
df.head(1000)
```

```python
df.to_csv('Preprocessing_imdb.csv',encoding='utf8', index=False)
stem_df = df[['stemming_data']]
stem_df.to_csv('stemming_imdb.csv',encoding='utf8', index=False)

stem_df = df[['comments']]
stem_df.to_csv('comments_imdb.csv',encoding='utf8', index=False)
```
